```python
# ─────────────────────────────────────────────────────────
#   IMPORTING REQUIRED LIBRARIES & DISPLAY UTILITIES
# ─────────────────────────────────────────────────────────

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from IPython.display import display, HTML

# ── Reusable styling utilities for the notebook ──────────

CARD = (
    "background: #0b1a2e;"
    "border-radius: 12px;"
    "padding: 28px 35px;"
    "margin: 10px 0;"
    "border: 1px solid rgba(45, 158, 174, 0.2);"
    "box-shadow: 0 4px 15px rgba(0, 0, 0, 0.3);"
    "font-family: 'Segoe UI', sans-serif;"
)

def render_card(title, body_html):
    display(HTML(f"""
    <div style="{CARD}">
        <p style="color: #4de0f5; font-size: 11px; letter-spacing: 4px;
                  text-transform: uppercase; margin: 0 0 16px 0; font-weight: 600;">{title}</p>
        <hr style="border: none; border-top: 1px solid rgba(92, 200, 216, 0.10); margin: 0 0 18px 0;">
        {body_html}
    </div>
    """))

def metric_row(label, value, highlight=False):
    val_color = "#4de0f5" if highlight else "#ffffff"
    return f"""
    <div style="display: flex; justify-content: space-between; align-items: center;
                padding: 11px 0; border-bottom: 1px solid rgba(45, 158, 174, 0.07);">
        <span style="color: #a8d8e8; font-size: 13px;">{label}</span>
        <span style="color: {val_color}; font-size: 14px; font-weight: 600;">{value}</span>
    </div>"""

def status_bar(text, sub=""):
    sub_html = f'<span style="color:#5ba8b5; font-size:12px; margin-left:20px;">{sub}</span>' if sub else ""
    display(HTML(f"""
    <div style="background: #0b1a2e; padding: 14px 24px; border-radius: 8px;
                border: 1px solid rgba(45, 158, 174, 0.12);
                border-left: 3px solid #2d8e9e; margin: 5px 0;">
        <span style="color: #4de0f5; font-family: 'Segoe UI'; font-size: 13px;">
            {text}</span>{sub_html}
    </div>"""))

def style_dataframe(dataframe):
    return (dataframe.style
        .set_table_styles([
            {'selector': 'th', 'props': [
                ('background-color', '#0d2137'), ('color', '#4de0f5'),
                ('font-family', "'Segoe UI', sans-serif"), ('font-size', '12px'),
                ('font-weight', '600'), ('padding', '12px 18px'),
                ('border', 'none'), ('border-bottom', '2px solid rgba(45,158,174,0.25)'),
                ('text-align', 'left'), ('letter-spacing', '0.5px')]},
            {'selector': 'td', 'props': [
                ('color', '#ffffff'), ('font-family', "'Segoe UI', sans-serif"),
                ('font-size', '12px'), ('padding', '10px 18px'),
                ('border', 'none'), ('border-bottom', '1px solid rgba(45,158,174,0.06)')]},
            {'selector': 'tr:nth-child(odd) td', 'props': [('background-color', '#0b1a2e')]},
            {'selector': 'tr:nth-child(even) td', 'props': [('background-color', '#0e2238')]},
            {'selector': 'tr:hover td', 'props': [('background-color', '#132d4a')]},
            {'selector': '', 'props': [
                ('border-collapse', 'collapse'),
                ('border', '1px solid rgba(45,158,174,0.15)'),
                ('border-radius', '10px'), ('overflow', 'hidden')]},
        ])
    )

# ── Confirmation ─────────────────────────────────────────

status_bar("Libraries imported successfully", f"pandas {pd.__version__}")
```

```python
# ─────────────────────────────────────────────────────────
#   DATA QUALITY ASSESSMENT
# ─────────────────────────────────────────────────────────

total_cells = df.shape[0] * df.shape[1]
total_missing = df.isnull().sum().sum()
total_duplicates = df.duplicated().sum()
completeness = ((total_cells - total_missing) / total_cells) * 100

# ── Summary metrics ──────────────────────────────────────

summary = (
    metric_row("Total Cells in Dataset", f"{total_cells:,}")
    + metric_row("Missing Values", f"{total_missing:,}", highlight=(total_missing > 0))
    + metric_row("Duplicate Rows", f"{total_duplicates:,}", highlight=(total_duplicates > 0))
    + metric_row("Data Completeness", f"{completeness:.1f}%")
)

render_card("Data Quality Assessment", summary)

# ── Per-column breakdown ─────────────────────────────────

col_rows = ""
for col, count in df.isnull().sum().items():
    if count == 0:
        badge = '<span style="color:#2d8e9e; font-size:11px; font-weight:600; letter-spacing:1px;">PASS</span>'
        val_text = "None detected"
    else:
        badge = '<span style="color:#ff6b6b; font-size:11px; font-weight:600; letter-spacing:1px;">FLAG</span>'
        val_text = f"{count} missing"
    col_rows += f"""
    <div style="display: flex; justify-content: space-between; align-items: center;
                padding: 10px 0; border-bottom: 1px solid rgba(45, 158, 174, 0.06);">
        <span style="color: #ffffff; font-size: 13px; width: 180px;">{col}</span>
        <span style="color: #7ec8d8; font-size: 13px; flex: 1;">{val_text}</span>
        {badge}
    </div>"""

render_card("Missing Values per Column", col_rows)
```

```python
# ─────────────────────────────────────────────────────────
#   DATA QUALITY ASSESSMENT
# ─────────────────────────────────────────────────────────

total_cells = df.shape[0] * df.shape[1]
total_missing = df.isnull().sum().sum()
total_duplicates = df.duplicated().sum()
completeness = ((total_cells - total_missing) / total_cells) * 100

print("┌──────────────────────────────────────────────────┐")
print("│          DATA QUALITY ASSESSMENT                 |")
print("├──────────────────────────────────────────────────┤")
print(f"│  Total Cells in Dataset:  {total_cells:<22} │")
print(f"│  Missing Values:          {total_missing:<22} │")
print(f"│  Duplicate Rows:          {total_duplicates:<22} │")
print(f"│  Data Completeness:       {completeness:<21.1f}% │")
print("└──────────────────────────────────────────────────┘")

print("\n  Missing Values per Column:")
print("  " + "─" * 48)
for col, count in df.isnull().sum().items():
    status = "None detected" if count == 0 else f"{count} missing"
    marker = "  [PASS]" if count == 0 else "  [FLAG]"
    print(f"   {col:<18} {status:<16}{marker}")
```

```python
# ─────────────────────────────────────────────────────────
#   PERFORMING DATA CLEANING OPERATIONS
# ─────────────────────────────────────────────────────────

rows_before = df.shape[0]
df = df.drop_duplicates()
rows_after_dedup = df.shape[0]
duplicates_removed = rows_before - rows_after_dedup

df = df.dropna()
rows_after_na = df.shape[0]
na_removed = rows_after_dedup - rows_after_na

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
date_dtype = str(df['Order_Date'].dtype)

# ── Step cards ───────────────────────────────────────────

def step_block(num, title, rows):
    rows_html = "".join([metric_row(l, v) for l, v in rows])
    return f"""
    <div style="margin-bottom: 22px;">
        <div style="display: flex; align-items: center; gap: 12px; margin-bottom: 10px;">
            <span style="background: #1b6b7d; color: #ffffff; font-size: 11px; font-weight: 700;
                         padding: 4px 10px; border-radius: 4px; letter-spacing: 1px;">STEP {num}</span>
            <span style="color: #ffffff; font-size: 14px; font-weight: 500;">{title}</span>
        </div>
        {rows_html}
    </div>"""

body = (
    step_block(1, "Remove Duplicates", [
        ("Rows before", f"{rows_before:,}"),
        ("Duplicates removed", f"{duplicates_removed:,}"),
        ("Rows after", f"{rows_after_dedup:,}"),
    ])
    + step_block(2, "Handle Missing Values", [
        ("Rows before", f"{rows_after_dedup:,}"),
        ("Rows dropped", f"{na_removed:,}"),
        ("Rows after", f"{rows_after_na:,}"),
    ])
    + step_block(3, "Convert Order_Date to DateTime", [
        ("Previous type", "object"),
        ("Converted to", date_dtype),
    ])
)

render_card("Cleaning Operations Log", body)
```

```python
# ─────────────────────────────────────────────────────────
#   POST-CLEANING VERIFICATION
# ─────────────────────────────────────────────────────────

# ── Summary metrics ──────────────────────────────────────

verify = (
    metric_row("Final Row Count", f"{df.shape[0]:,}")
    + metric_row("Final Column Count", f"{df.shape[1]}")
    + metric_row("Remaining Missing Values", f"{df.isnull().sum().sum()}")
    + metric_row("Remaining Duplicates", f"{df.duplicated().sum()}")
)

# ── Data types ───────────────────────────────────────────

dtype_rows = ""
for col, dtype in df.dtypes.items():
    dtype_rows += f"""
    <div style="display: flex; justify-content: space-between; padding: 9px 0;
                border-bottom: 1px solid rgba(45, 158, 174, 0.06);">
        <span style="color: #ffffff; font-size: 13px;">{col}</span>
        <span style="color: #7ec8d8; font-size: 13px;">{dtype}</span>
    </div>"""

status_badge = """
<div style="margin-top: 22px; padding: 14px 20px; background: rgba(45, 158, 174, 0.08);
            border-radius: 8px; border: 1px solid rgba(45, 158, 174, 0.15); text-align: center;">
    <span style="color: #4de0f5; font-size: 13px; font-weight: 600; letter-spacing: 3px;
                 text-transform: uppercase;">Dataset is clean and ready for analysis</span>
</div>"""

body = verify + f"""
<p style="color: #4de0f5; font-size: 11px; letter-spacing: 3px; text-transform: uppercase;
          font-weight: 600; margin: 22px 0 10px 0;">Final Data Types</p>
<hr style="border: none; border-top: 1px solid rgba(92, 200, 216, 0.08); margin: 0 0 6px 0;">
{dtype_rows}
{status_badge}
"""

render_card("Clean Dataset Verification", body)
```

```python
# ─────────────────────────────────────────────────────────
#   KPI (a): TOTAL REVENUE
# ─────────────────────────────────────────────────────────

total_revenue = df['Revenue'].sum()
total_orders = df.shape[0]
avg_revenue_per_order = df['Revenue'].mean()

body = f"""
<p style="color: #5ba8b5; font-size: 11px; letter-spacing: 3px; text-transform: uppercase;
          margin: 0 0 6px 0;">Total Revenue Generated</p>
<p style="color: #ffffff; font-size: 38px; font-weight: 300; margin: 0 0 24px 0;
          letter-spacing: 1px;">R{total_revenue:,.2f}</p>
<hr style="border: none; border-top: 1px solid rgba(92, 200, 216, 0.08); margin: 0 0 14px 0;">
{metric_row("Total Orders Processed", f"{total_orders:,}")}
{metric_row("Average Revenue per Order", f"R{avg_revenue_per_order:,.2f}")}
"""

render_card("KPI (a)  —  Total Revenue", body)
```

```python
# ─────────────────────────────────────────────────────────
#   KPI (b): AVERAGE UNITS SOLD PER ORDER
# ─────────────────────────────────────────────────────────

avg_units = df['Units_Sold'].mean()
min_units = df['Units_Sold'].min()
max_units = df['Units_Sold'].max()
median_units = df['Units_Sold'].median()

body = f"""
<p style="color: #5ba8b5; font-size: 11px; letter-spacing: 3px; text-transform: uppercase;
          margin: 0 0 6px 0;">Average Units per Order</p>
<p style="color: #ffffff; font-size: 38px; font-weight: 300; margin: 0 0 24px 0;
          letter-spacing: 1px;">{avg_units:,.2f}</p>
<hr style="border: none; border-top: 1px solid rgba(92, 200, 216, 0.08); margin: 0 0 14px 0;">
{metric_row("Median Units per Order", f"{median_units:,.2f}")}
{metric_row("Minimum in a Single Order", f"{min_units:,}")}
{metric_row("Maximum in a Single Order", f"{max_units:,}")}
"""

render_card("KPI (b)  —  Average Units Sold", body)
```

```python
# ─────────────────────────────────────────────────────────
#   KPI (c): TOTAL REVENUE PER REGION
# ─────────────────────────────────────────────────────────

revenue_by_region = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
max_rev = revenue_by_region.max()

region_rows = ""
for rank, (region, revenue) in enumerate(revenue_by_region.items(), 1):
    pct = (revenue / max_rev) * 100
    if rank == 1:
        tag = '<span style="color:#4de0f5; font-size:10px; font-weight:600; letter-spacing:1px; margin-left:10px;">HIGHEST</span>'
    elif rank == len(revenue_by_region):
        tag = '<span style="color:#5ba8b5; font-size:10px; font-weight:600; letter-spacing:1px; margin-left:10px;">LOWEST</span>'
    else:
        tag = ""

    region_rows += f"""
    <div style="padding: 12px 0; border-bottom: 1px solid rgba(45, 158, 174, 0.06);">
        <div style="display: flex; justify-content: space-between; align-items: center; margin-bottom: 6px;">
            <span style="color: #ffffff; font-size: 13px; font-weight: 500;">
                <span style="color: #5ba8b5; font-size: 12px; margin-right: 10px;">{rank}.</span>{region}{tag}
            </span>
            <span style="color: #4de0f5; font-size: 13px; font-weight: 600;">R{revenue:,.2f}</span>
        </div>
        <div style="background: rgba(45, 158, 174, 0.08); border-radius: 3px; height: 5px;">
            <div style="background: linear-gradient(90deg, #1b6b7d, #4de0f5);
                        border-radius: 3px; height: 5px; width: {pct:.0f}%;"></div>
        </div>
    </div>"""

total_row = f"""
<div style="display: flex; justify-content: space-between; padding: 16px 0 0 0;">
    <span style="color: #4de0f5; font-size: 13px; font-weight: 600; letter-spacing: 2px;
                 text-transform: uppercase;">Total</span>
    <span style="color: #ffffff; font-size: 15px; font-weight: 700;">R{revenue_by_region.sum():,.2f}</span>
</div>"""

render_card("KPI (c)  —  Total Revenue per Region", region_rows + total_row)
```

```python
# ─────────────────────────────────────────────────────────
#   KPI (d): HIGHEST REVENUE SALES REPRESENTATIVE
# ─────────────────────────────────────────────────────────

revenue_by_rep = df.groupby('Sales_Rep')['Revenue'].sum().sort_values(ascending=False)
top_rep = revenue_by_rep.index[0]
top_rep_revenue = revenue_by_rep.values[0]
max_rep_rev = revenue_by_rep.max()

# ── Top performer highlight ──────────────────────────────

highlight = f"""
<div style="background: rgba(45, 158, 174, 0.08); border-radius: 10px;
            padding: 20px 28px; margin-bottom: 22px;
            border: 1px solid rgba(45, 158, 174, 0.15);">
    <p style="color: #5ba8b5; font-size: 11px; letter-spacing: 3px;
              text-transform: uppercase; margin: 0 0 6px 0;">Top Performer</p>
    <div style="display: flex; justify-content: space-between; align-items: baseline;">
        <span style="color: #ffffff; font-size: 28px; font-weight: 300;
                     letter-spacing: 1px;">{top_rep}</span>
        <span style="color: #4de0f5; font-size: 24px; font-weight: 300;">R{top_rep_revenue:,.2f}</span>
    </div>
</div>"""

# ── Full rankings ────────────────────────────────────────

rep_rows = ""
for rank, (rep, revenue) in enumerate(revenue_by_rep.items(), 1):
    pct = (revenue / max_rep_rev) * 100
    is_top = rank == 1
    name_color = "#4de0f5" if is_top else "#ffffff"
    val_weight = "700" if is_top else "500"

    rep_rows += f"""
    <div style="padding: 10px 0; border-bottom: 1px solid rgba(45, 158, 174, 0.06);">
        <div style="display: flex; justify-content: space-between; align-items: center; margin-bottom: 5px;">
            <span style="color: {name_color}; font-size: 13px;">
                <span style="color: #5ba8b5; font-size: 12px; margin-right: 10px; display: inline-block; width: 22px; text-align: right;">{rank}.</span>{rep}
            </span>
            <span style="color: #ffffff; font-size: 13px; font-weight: {val_weight};">R{revenue:,.2f}</span>
        </div>
        <div style="background: rgba(45, 158, 174, 0.08); border-radius: 3px; height: 4px;">
            <div style="background: linear-gradient(90deg, #1b6b7d, #4de0f5);
                        border-radius: 3px; height: 4px; width: {pct:.0f}%;"></div>
        </div>
    </div>"""

body = highlight + f"""
<p style="color: #4de0f5; font-size: 11px; letter-spacing: 3px; text-transform: uppercase;
          font-weight: 600; margin: 0 0 10px 0;">Complete Rankings</p>
<hr style="border: none; border-top: 1px solid rgba(92, 200, 216, 0.08); margin: 0 0 6px 0;">
{rep_rows}"""

render_card("KPI (d)  —  Sales Representative Rankings", body)
```

```python
# ─────────────────────────────────────────────────────────
#   KPI (e): TOP 3 PRODUCTS BY TOTAL UNITS SOLD
# ─────────────────────────────────────────────────────────

units_by_product = df.groupby('Product')['Units_Sold'].sum().sort_values(ascending=False)
max_units_val = units_by_product.max()

product_rows = ""
for rank, (product, units) in enumerate(units_by_product.items(), 1):
    pct = (units / max_units_val) * 100
    is_top3 = rank <= 3

    if is_top3:
        name_color = "#4de0f5"
        val_color = "#4de0f5"
        badge = f"""<span style="background: rgba(45, 158, 174, 0.15); color: #4de0f5;
                     font-size: 9px; font-weight: 700; padding: 2px 8px;
                     border-radius: 3px; letter-spacing: 1px; margin-left: 10px;">TOP 3</span>"""
        bar_gradient = "linear-gradient(90deg, #1b6b7d, #4de0f5)"
    else:
        name_color = "#ffffff"
        val_color = "#7ec8d8"
        badge = ""
        bar_gradient = "linear-gradient(90deg, rgba(27,107,125,0.4), rgba(77,224,245,0.3))"

    product_rows += f"""
    <div style="padding: 12px 0; border-bottom: 1px solid rgba(45, 158, 174, 0.06);">
        <div style="display: flex; justify-content: space-between; align-items: center; margin-bottom: 6px;">
            <span style="color: {name_color}; font-size: 13px; font-weight: 500;">
                <span style="color: #5ba8b5; font-size: 12px; margin-right: 10px;
                             display: inline-block; width: 22px; text-align: right;">{rank}.</span>{product}{badge}
            </span>
            <span style="color: {val_color}; font-size: 13px; font-weight: 600;">{units:,} units</span>
        </div>
        <div style="background: rgba(45, 158, 174, 0.08); border-radius: 3px; height: 5px;">
            <div style="background: {bar_gradient};
                        border-radius: 3px; height: 5px; width: {pct:.0f}%;"></div>
        </div>
    </div>"""

total_row = f"""
<div style="display: flex; justify-content: space-between; padding: 16px 0 0 0;">
    <span style="color: #4de0f5; font-size: 13px; font-weight: 600; letter-spacing: 2px;
                 text-transform: uppercase;">Total Units Sold</span>
    <span style="color: #ffffff; font-size: 15px; font-weight: 700;">{units_by_product.sum():,} units</span>
</div>"""

render_card("KPI (e)  —  Product Rankings by Units Sold", product_rows + total_row)
```

```python
# ─────────────────────────────────────────────────────────
#   CONFIGURING VISUALIZATION ENVIRONMENT
# ─────────────────────────────────────────────────────────

import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

plt.rcParams['figure.facecolor'] = '#0a1628'
plt.rcParams['axes.facecolor'] = '#0b1a2e'
plt.rcParams['text.color'] = '#ffffff'
plt.rcParams['axes.labelcolor'] = '#4de0f5'
plt.rcParams['xtick.color'] = '#ffffff'
plt.rcParams['ytick.color'] = '#ffffff'
plt.rcParams['font.family'] = 'Segoe UI'
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.color'] = '#1b7a8a'
plt.rcParams['grid.alpha'] = 0.15

print("  Visualization environment configured")
print("  Premium theme applied")
```

```python
# ─────────────────────────────────────────────────────────
#   CHART 1: TOTAL REVENUE BY REGION
# ─────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(14, 7))

colors = ['#0c3547', '#0f4d62', '#14667a', '#1b8a9a', '#26a8b8', '#3bc4d2', '#4de0f5']
revenue_sorted = revenue_by_region.sort_values(ascending=True)

bars = ax.barh(
    revenue_sorted.index,
    revenue_sorted.values,
    color=colors,
    edgecolor=(0.30, 0.88, 0.96, 0.3),
    linewidth=0.8,
    height=0.55
)

for bar, value in zip(bars, revenue_sorted.values):
    ax.text(
        value + 80000,
        bar.get_y() + bar.get_height()/2,
        f'R{value:,.0f}',
        va='center',
        fontsize=11,
        color='#4de0f5',
        fontweight='bold'
    )

ax.set_title('Total Revenue by Region', fontsize=22, fontweight='bold', color='#ffffff', pad=30)
ax.set_xlabel('Revenue (R)', fontsize=13, labelpad=15, color='#4de0f5')
ax.set_ylabel('')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: f'R{x:,.0f}'))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color((0.18, 0.62, 0.68, 0.25))
ax.spines['left'].set_color((0.18, 0.62, 0.68, 0.25))

plt.tight_layout()
plt.show()
```

```python
# ─────────────────────────────────────────────────────────
#   CHART 2: TOTAL REVENUE BY PRODUCT
# ─────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(14, 7))

revenue_by_product = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
colors = ['#4de0f5', '#3bc4d2', '#26a8b8', '#1b8a9a', '#14667a', '#0f4d62', '#0c3547']

bars = ax.bar(
    revenue_by_product.index,
    revenue_by_product.values,
    color=colors,
    edgecolor=(0.30, 0.88, 0.96, 0.3),
    linewidth=0.8,
    width=0.55
)

for bar, value in zip(bars, revenue_by_product.values):
    ax.text(
        bar.get_x() + bar.get_width()/2,
        value + 100000,
        f'R{value:,.0f}',
        ha='center',
        fontsize=10,
        color='#4de0f5',
        fontweight='bold'
    )

ax.set_title('Total Revenue by Product', fontsize=22, fontweight='bold', color='#ffffff', pad=30)
ax.set_xlabel('Product', fontsize=13, labelpad=15, color='#4de0f5')
ax.set_ylabel('Revenue (R)', fontsize=13, labelpad=15, color='#4de0f5')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: f'R{x:,.0f}'))
plt.xticks(rotation=20, ha='right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color((0.18, 0.62, 0.68, 0.25))
ax.spines['left'].set_color((0.18, 0.62, 0.68, 0.25))

plt.tight_layout()
plt.show()
```

```python
# ─────────────────────────────────────────────────────────
#   CHART 3: MONTHLY REVENUE TREND
# ─────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(14, 7))

df['Month'] = df['Order_Date'].dt.month
monthly_revenue = df.groupby('Month')['Revenue'].sum()
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

ax.fill_between(
    range(len(monthly_revenue)),
    monthly_revenue.values,
    alpha=0.12,
    color='#4de0f5'
)

ax.plot(
    range(len(monthly_revenue)),
    monthly_revenue.values,
    color='#4de0f5',
    linewidth=2.5,
    marker='o',
    markersize=10,
    markerfacecolor='#4de0f5',
    markeredgecolor='#ffffff',
    markeredgewidth=1.5
)

for i, value in enumerate(monthly_revenue.values):
    ax.text(
        i, value + 200000,
        f'R{value:,.0f}',
        ha='center',
        fontsize=11,
        color='#4de0f5',
        fontweight='bold'
    )

ax.set_title('Monthly Revenue Trend  —  2024', fontsize=22, fontweight='bold', color='#ffffff', pad=30)
ax.set_xlabel('Month', fontsize=13, labelpad=15, color='#4de0f5')
ax.set_ylabel('Revenue (R)', fontsize=13, labelpad=15, color='#4de0f5')
ax.set_xticks(range(len(month_names)))
ax.set_xticklabels(month_names, fontsize=13)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: f'R{x:,.0f}'))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color((0.18, 0.62, 0.68, 0.25))
ax.spines['left'].set_color((0.18, 0.62, 0.68, 0.25))

plt.tight_layout()
plt.show()
```

```python
# ─────────────────────────────────────────────────────────
#   EXECUTIVE SUMMARY
# ─────────────────────────────────────────────────────────

top_region = revenue_by_region.index[0]
top_region_rev = revenue_by_region.values[0]
low_region = revenue_by_region.index[-1]
low_region_rev = revenue_by_region.values[-1]
top_product_rev = revenue_by_product.index[0]
top_product_units = units_by_product.index[0]

# ── KPI tiles row ────────────────────────────────────────

def kpi_tile(label, value):
    return f"""
    <div style="flex: 1; background: rgba(45, 158, 174, 0.06); border-radius: 10px;
                padding: 18px 22px; border: 1px solid rgba(45, 158, 174, 0.1);">
        <p style="color: #5ba8b5; font-size: 10px; letter-spacing: 3px;
                  text-transform: uppercase; margin: 0 0 6px 0;">{label}</p>
        <p style="color: #ffffff; font-size: 20px; font-weight: 300; margin: 0;
                  letter-spacing: 0.5px;">{value}</p>
    </div>"""

tiles = f"""
<div style="display: flex; gap: 16px; margin-bottom: 28px;">
    {kpi_tile("Total Revenue", f"R{total_revenue:,.2f}")}
    {kpi_tile("Total Orders", f"{total_orders:,}")}
    {kpi_tile("Avg Revenue / Order", f"R{avg_revenue_per_order:,.2f}")}
    {kpi_tile("Avg Units / Order", f"{avg_units:,.2f}")}
</div>"""

# ── Sections ─────────────────────────────────────────────

def section_title(text):
    return f"""
    <p style="color: #4de0f5; font-size: 11px; letter-spacing: 3px; text-transform: uppercase;
              font-weight: 600; margin: 24px 0 10px 0;">{text}</p>
    <hr style="border: none; border-top: 1px solid rgba(92, 200, 216, 0.08); margin: 0 0 8px 0;">"""

geographic = (
    section_title("Geographic Insights")
    + metric_row("Highest Performing Region", f"{top_region}  —  R{top_region_rev:,.2f}", highlight=True)
    + metric_row("Lowest Performing Region", f"{low_region}  —  R{low_region_rev:,.2f}")
)

product = (
    section_title("Product Insights")
    + metric_row("Top Product by Revenue", top_product_rev, highlight=True)
    + metric_row("Top Product by Volume", top_product_units, highlight=True)
)

sales_team = (
    section_title("Sales Team")
    + metric_row("Top Performer", f"{top_rep}  —  R{top_rep_revenue:,.2f}", highlight=True)
)

key_insight = f"""
{section_title("Key Insight")}
<div style="background: rgba(45, 158, 174, 0.06); border-radius: 8px;
            padding: 18px 22px; margin-top: 8px;
            border: 1px solid rgba(45, 158, 174, 0.1);
            border-left: 3px solid #4de0f5;">
    <p style="color: #ffffff; font-size: 13px; margin: 0; line-height: 1.7; font-family: 'Segoe UI';">
        The highest-selling product by <span style="color: #4de0f5; font-weight: 600;">units</span>
        ({top_product_units}) differs from the highest by
        <span style="color: #4de0f5; font-weight: 600;">revenue</span> ({top_product_rev}),
        indicating premium pricing on {top_product_rev}s drives more revenue despite lower volume.
    </p>
</div>"""

body = tiles + geographic + product + sales_team + key_insight

render_card("Executive Summary", body)
```
